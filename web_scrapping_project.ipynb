## Web Scraping and Analaysis on Urban Comapny

### Problem Statement:-
- Scrapping the data from urban company website data like reviews, ratings services, country, city.
- Link for the site:- https://www.urbancompany.com/reviews this link covers only the reviews by the customers.

```python
#Importing important libraries.
from bs4 import BeautifulSoup   # BS4 for Scrapping data from site.
import requests                 # Request to get the url
import pandas as pd      
import numpy as np# Pandas to create DataFrame 
```

```python
url = "https://www.urbancompany.com/reviews"  # link contains Reviews only 
s = requests.get(url)                         
soup = BeautifulSoup(s.content, 'html.parser')               
```

```python
com = soup.find("div", class_ = "ReviewCollectionParent__reviewActualPane--1dQI4")
```

```python
com
```

```python
names = []
dates = []
rates = []
comments = []
services = []
for i in range(1, 1000): #Loop for multiple pages.
    
    urls = f"https://www.urbancompany.com/reviews?p={i}" 
    ct = requests.get(urls)
    soup = BeautifulSoup(ct.content)


    coms = soup.find_all("div", class_ ="ReviewPageReview__topInfo--3ibra")


    for com in coms:
        name = com.find("span", class_ = "ellipsis").text 
        date = com.find("span", class_ = "ReviewPageReview__reviewDate--3tC00").text    
        ratings = com.find("div", class_ = "StarRatingSEO__ratingBox--3MqzX StarRatingSEO__ratingBoxSmall--4Nhn8").text
        reviews = com.find("div", class_ = "ReviewPageReview__review--1MFrH ReviewPageReview__desktopOnly--MXldJ")
        service = com.find("span",class_="ReviewPageReview__seoLink--28aIY ellipsis")
        if reviews is None:
            comments.append("None")
        else:
            comments.append(reviews.text)
        if service is None:
            services.append("None")
        else:
            services.append(service.text)

        names.append(name)
        dates.append(date)
        rates.append(ratings)
    
        
#         print(f"Name : {name}")
#         print(f"Date : {date}")
#         print(f"Ratings : {ratings}")
#         print("="*50)
```

```python
#Extracting the services and locations from 'services' field
Services = []
Places = []

for i in services:
    
    if i != "None":
        services,place=i.split(" in ")
        Services.append(services)
        Places.append(place)
        
    elif i == "None":
        Services.append("None")
        Places.append("None")
```

```python
#Extractiing Country and City names from the previously extracted "Places" column
 Country = []
City = []
for i in Places:
    
    if i == "None":
        City.append("None")
        Country.append("None")
        
    elif i != "None":
        city,country=i.split(", ")
        Country.append(country)
        City.append(city)
```

```python
#Creating empty dataframe
df = pd.DataFrame()
```

```python
df["Dates"] = dates
df["Names"] = names
df["Ratings"] = rates
df["Services"] = Services
df["City"] = City
df["Country"] = Country
df["Comments"] = comments
```
